# Orders Table

Let's start by preparing the orders table: Let's find out the total value per order and the nr of products per order.

```python
order_value = orders_data.groupby("Order ID").Sales.agg("sum")
order_nr_of_products = orders_data.groupby("Order ID").Sales.agg("count")
```

```python
orders_data = orders_data.drop(columns = ["Sales"])
orders_data = orders_data.drop_duplicates(subset='Order ID', keep='first')
```

```python
orders_data = pd.merge(orders_data, order_value,
how='left', on="Order ID")
orders_data = pd.merge(orders_data, order_nr_of_products,
how='left', on="Order ID")
orders_data = orders_data.rename(columns = {"Sales_x":"Revenue", "Sales_y":'Nr_of_products'})
```

Now, after organizing and merging let's have a look on the current state of the table

```python
orders_data.head()
```

Now let's prepare some columns that are a result of other columns operations and that will be usefull later on.
 
 -Duration of shipping
 
 -Average price per order

```python
orders_data["shipment_duration"] = orders_data["Ship Date"] - orders_data["Order Date"]
orders_data["Mean prod value"] = orders_data["Revenue"] / orders_data["Nr_of_products"]
```

I also want a table that demostrates sales per day per region. The first thing to do is to know what regions we have.


```python
orders_data["Region"].value_counts()
```

Now let's filter the dataframes by region and create time series of sales and nr of sold items in each one.


```python
#create a custome fucntion to crunch the columns

def region_separator( df , region ):
    try:
        output = df[df.Region == region]
    except:
        print("Wrong inputs!")
    
    region_revenue = output.groupby("Order Date").Revenue.agg("sum")
    region_sales = output.groupby("Order Date").Nr_of_products.agg("sum")
    
    output = pd.concat([region_revenue,region_sales], axis = 1)
    
    
    output = output.rename(columns = {"Revenue" :  "R_"+ region ,"Nr_of_products" : "Vol_" + region  })
    return output

orders_west = region_separator(orders_data,"West")
orders_east = region_separator(orders_data,"East")
orders_north = region_separator(orders_data,"North")
orders_south = region_separator(orders_data,"South")

aux =pd.merge(orders_west,orders_east, how = "outer" , on= "Order Date")  
aux =pd.merge(aux,orders_north, how = "outer" , on= "Order Date")  
aux =pd.merge(aux,orders_south, how = "outer" , on= "Order Date") 
print(aux.isnull().sum())

region_time_series = aux.fillna(0)
#orders_south
region_time_series

%store region_time_series
```
